FLOWER CLASSIFICATION USING TENSOR FLOW AND TRAnsfer learning

```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import tensorflow_datasets as tfds
```

```python
(train_ds,train_labels),(test_ds,test_labels) = tfds.load(
    "tf_flowers",
    split=["train[:70%]", "train[70%:]"],
    batch_size=-1,
    as_supervised=True,

)
```

```python
train_ds[0].shape
```

```python
train_ds=tf.image.resize(train_ds,(150,150))
test_ds=tf.image.resize(test_ds,(150,150))
```

```python
train_labels
```

```python
trian_labels= to_categorical(train_labels,num_classes=5)
test_labels= to_categorical(test_labels,num_classes=5)
```

```python
trian_labels[0]
```

USING PRE-TRAINED VGG16 IMAGE CLASSFICATION

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
base_model=VGG16(weights="imagenet",include_top=False,input_shape=train_ds[0].shape)
```

```python
base_model.trainable=False
```

```python
train_ds=preprocess_input(train_ds)
test_ds=preprocess_input(test_ds)
```

```python
base_model.summary()
```

```python
from tensorflow.keras import layers, models

# Define layers
flatten_layer = layers.Flatten()
dense_layer_1 = layers.Dense(50, activation="relu")
dense_layer_2 = layers.Dense(20, activation="relu")
prediction_layer = layers.Dense(5, activation="softmax")

# Initialize the model with the correct Sequential syntax
model = models.Sequential([
    base_model,
    flatten_layer,
    dense_layer_1,
    dense_layer_2,
    prediction_layer
])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
es=EarlyStopping(monitor="val_accuracy",mode="max",patience=5, restore_best_weights=True)
```

```python
model.fit(train_ds,trian_labels,epochs=50,validation_split=0.2,batch_size=32,callbacks=[es])
```

```python
loss, accuracy = model.evaluate(test_ds, test_labels)
print(f'Test Accuracy: {accuracy:.2f}')
```

```python
model.save('flower_classification_model.h5')
```
